# Exercise 3 (Timing Stuff)

```python
squares = [i**2 for i in range(10)] 
print(squares)
%timeit for i in range(10): squares
```

Analysis: 
- 356ns is the time taken to 1 loop to finish, with an uncertainty of 41.8ns per run.
- the 41.8ns uncertainty is the standard deviation of 7,000,000 loops (7 runs x 1,000,000 loops)

```python
squares = list()
for i in range(10):
    squares.append(i**2)
print (squares)
%timeit for i in range(10): squares
```

Analysis:
- 335ns is the time taken for 1 loop to finish, with an uncertainty of 35.2ns per run.
- the 35.2ns uncertainty is the standard deviation of 7,000,000 loops (7 runs x 1,000,000 loops)

Further thoughts: 
- Why is it faster for the 'longer' code to process and output a result compared to the 'shorter' code? (335 < 356)
- Likewise, why is the 'longer' code processing and outputting at a more consistent rate as compared to the 'shorter' code? (35.2 < 41.8)
