# Exercise 3 (Timing Stuff)

```python
%%timeit
squares = [i**2 for i in range(10)] 
# print(squares)
# %timeit for i in range(10): squares
```

Analysis: 
- 1.35 μs is the time taken to 1 loop to finish, with an uncertainty of 26.1ns per run.
- the 26.1ns uncertainty is the standard deviation of 7,000,000 loops (7 runs x 1,000,000 loops)

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
# print (squares)
# %timeit for i in range(10): squares
```

Analysis:
- 1.78μs is the time taken for 1 loop to finish, with an uncertainty of 135ns per run.
- the 135ns uncertainty is the standard deviation of 7,000,000 loops (7 runs x 1,000,000 loops)

Futher thoughts:
- The 'append' in the second code causes it to have to perform an extra action, especially with more lines of code to process per cycle in the second round of code. Hence, having your code be more compact is better in general.
- However, it begs the question of whether there is the case where a single line of code processing will take longer than if it is split up into multiple lines.
